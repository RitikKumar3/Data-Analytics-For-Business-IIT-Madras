### DATA ANALYSIS AND VISUALIZATION

```python
# Show the countries with higher GDP in the year 2010
higher_gdp_countries_in_2010 = df[df['Year']==2010].sort_values(by='gdp_per_capita', ascending = False)[['Year', 'Entity', 'gdp_per_capita']][0:10]
higher_gdp_countries_in_2010
```

```python
#Plot ten countries with highest GDP in the year 2010
plt.figure(figsize=(14, 5))
sns.barplot(x= 'Entity', y = 'gdp_per_capita', data = higher_gdp_countries_in_2010).set(xlabel ="Country name", ylabel = "GDP per Capita ($)", title='Ten countries with highest GDP in the year 2010')
plt.show()
```

```python
#The GDP per Capita change in all years of Qatar (Example)
Qatar_gdp = df[df['Entity'] == 'Qatar'].sort_values(by='Year', ascending = True)[['Year', 'Entity', 'gdp_per_capita']]
Qatar_gdp.head()
```

```python
#Plot the GDP per Capita change in all years of Qatar (Example)
fig, ax = plt.subplots(figsize = (10, 5))
sns.lineplot(x= 'Year', y = 'gdp_per_capita', data = Qatar_gdp, marker = 'o', color = 'r').set(xlabel ="Year", ylabel = "GDP per Capita ($)", title='The GDP per Capita change in all years of Qatar')
ax.set_xticks(Qatar_gdp.Year[::4])
plt.show()
```

```python
#Change country names to ISO code
import country_converter as coco
ISO_code_name_countries = coco.convert(names = df['Entity'], to = 'ISO3')
df['Country_names_ISO3'] = ISO_code_name_countries
```

```python
#Map data to world map for Energy intensity level of primary energy (MJ/$2017 PPP GDP) in 2010
D_to_map = df[df.Year == 2010]['Energy intensity level of primary energy (MJ/$2017 PPP GDP)']
locations = df[df.Year == 2010]['Country_names_ISO3']
names = df[df.Year == 2010]['Entity']
```

```python
locations.shape == D_to_map.shape
```

```python
#Plot
go.Figure(data = go.Choropleth(
    locations = locations,
    z = D_to_map,
    text = names,
    colorscale = 'Blues',
    autocolorscale=True,
    reversescale=True,
    marker_line_color='darkgray',
    marker_line_width=0.5,
    colorbar_title = 'MJ/$2017 PPP GDP'
)).update_layout(title_text='Energy intensity level of primary energy in 2010', autosize=False, width=1000, height=500)
```
